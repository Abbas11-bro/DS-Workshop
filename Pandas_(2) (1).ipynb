# Groupby

The groupby method allows you to group rows of data together and call aggregate functions

```python
import pandas as pd
# Our data: dictionary of lists
data = {'Company':['Alvi','Alvi','Shanooz','Shanooz','Bilalz','Bilalz'],
       'Person':['Armaan','Sufian','Arzaan','Shazeb','Suhail','Arsh'],
       'Sales':[200,120,340,124,243,350]}
```

```python
df = pd.DataFrame(data)
```

```python
df
```

** Now you can use the .groupby() method to group rows together based off of a column name. For instance let's group based off of Company. This will create a DataFrameGroupBy object:**

```python
df.groupby('Company')
```

```python
df
```

```python
by_comp = df.groupby("Company")['Sales'].mean()

print(by_comp)
```

```python
by_comp = df.groupby("Company")['Sales'].sum()

print(by_comp)
```

```python
by_comp = df.groupby("Company")['Sales'].max()

print(by_comp)
```

```python
by_comp = df.groupby("Company")['Sales'].min()

print(by_comp)
```
